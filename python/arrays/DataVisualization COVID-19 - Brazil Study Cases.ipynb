```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
base_path = os.path.abspath("")
```

```python
df_covid = pd.read_csv("{}/covid19_sus_dataset_r0.csv".format(base_path), sep=";")
```

```python
df_covid
```

```python
display(df_covid.describe(), df_covid.columns)
```

```python
df_covid["data"] = pd.to_datetime(df_covid["data"])
```

```python
df_covid["mes_year"] = [ "{:02d}{}".format(x.month, x.year) for x in df_covid["data"]]
display(df_covid["mes_year"])
```

```python
df_datas = pd.DataFrame([x for x in df_covid["data"].unique()])
```

```python
#split the dataset by UF
df_covidbyuf = df_covid.groupby(['estado', 'mes_year'], as_index = False)[['obitosNovos','casosNovos']].agg({ 'obitosNovos' : ['mean'], 'casosNovos' : ['mean'] })
display(df_covidbyuf.head(120))
```

```python
uf_list = np.sort([ x for x in df_covid['estado'].unique()])

fig, axs2 = plt.subplots(figsize=(14, 10))

x_axis = [y for y in df_covid['mes_year'].unique()]

for i, f in enumerate(uf_list):
    #print(i, f)
    
    wh_by_estado = df_covidbyuf['estado'] == f
    wh_by_deny_zero = df_covidbyuf['obitosNovos'] > 0
    df = df_covidbyuf.where(wh_by_estado)[['mes_year', 'obitosNovos']].dropna()
    
    y_axis = df['obitosNovos']
    
    axs2.plot(x_axis, y_axis, label="{}".format(f))
    axs2.legend()
    
    #if i == 2:
    #    break
plt.xticks(rotation=45)
plt.xlabel("Data (Mes/Ano)")
plt.ylabel("Nr. de Obitos")
plt.title("Numero de Ã“bitos por data")
plt.grid(True)
plt.show()
```
